## Setting and Calling API

```python
import openmeteo_requests
import pandas as pd
import requests_cache
from retry_requests import retry

# Setup the Open-Meteo API client with cache and retry on error
cache_session = requests_cache.CachedSession('.cache', expire_after=3600)
retry_session = retry(cache_session, retries=5, backoff_factor=0.2)
openmeteo = openmeteo_requests.Client(session=retry_session)

# API setup
url = "https://api.open-meteo.com/v1/forecast"
variables = [
    "temperature_2m",
    "relative_humidity_2m",
    "dew_point_2m",
    "apparent_temperature",
    "precipitation",
    "windspeed_10m",
    "surface_pressure"
]

params = {
    "latitude": 51.5085,
    "longitude": -0.1257,
    "hourly": ",".join(variables),
    "timezone": "Europe/London"
}

# Fetch response
responses = openmeteo.weather_api(url, params=params)
response = responses[0]

# Location info
print(f"Coordinates: {response.Latitude()}°N {response.Longitude()}°E")
print(f"Elevation: {response.Elevation()} m asl")
print(f"Timezone: {response.Timezone()} {response.TimezoneAbbreviation()}")
print(f"UTC offset: {response.UtcOffsetSeconds()} seconds")

# Create hourly dataframe
hourly = response.Hourly()
hourly_data = {
    "date": pd.date_range(
        start=pd.to_datetime(hourly.Time(), unit="s", utc=True),
        end=pd.to_datetime(hourly.TimeEnd(), unit="s", utc=True),
        freq=pd.Timedelta(seconds=hourly.Interval()),
        inclusive="left"
    )
}

# Add each variable
for i, var in enumerate(variables):
    hourly_data[var] = hourly.Variables(i).ValuesAsNumpy()

# Create DataFrame
hourly_df = pd.DataFrame(hourly_data)
print("\nLondon Hourly Weather Data:")
print(hourly_df.head())
```

## Data Exploration 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
plt.rcParams["axes.labelsize"] = 12
plt.rcParams["axes.titlesize"] = 14

# Variables to plot
variables_to_plot = [
    "temperature_2m",
    "relative_humidity_2m",
    "dew_point_2m",
    "apparent_temperature",
    "precipitation",
    "windspeed_10m",
    "surface_pressure"
]

# 1. Line plots for all variables
for var in variables_to_plot:
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=hourly_df, x="date", y=var)
    plt.title(f"Hourly {var.replace('_', ' ').title()} in London")
    plt.xlabel("Date & Time")
    plt.ylabel(var.replace("_", " ").title())
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"plots/{var}_line_plot_Ln.png")  # Save in 'plots/' dir (make sure it exists)
    plt.show()

# 2. Rolling averages for selected variables
rolling_vars = ["temperature_2m", "windspeed_10m", "relative_humidity_2m"]
window_size = 6  # 6-hour rolling average

for var in rolling_vars:
    hourly_df[f"{var}_rolling"] = hourly_df[var].rolling(window=window_size).mean()

    plt.figure(figsize=(12, 6))
    sns.lineplot(x="date", y=var, data=hourly_df, label="Actual", alpha=0.5)
    sns.lineplot(x="date", y=f"{var}_rolling", data=hourly_df, label=f"{window_size}-hr Rolling Avg", color="green")
    plt.title(f"{var.replace('_', ' ').title()} with {window_size}-Hour Rolling Average")
    plt.xlabel("Time")
    plt.ylabel(var.replace("_", " ").title())
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"plots/{var}_rolling_avg_Ln.png")
    plt.show()
```

```python
# 3. Summary statistics
print("\n--- Summary Statistics ---")
print(hourly_df[variables_to_plot].describe())

# 4. Correlation heatmap
plt.figure(figsize=(10, 6))
corr = hourly_df[variables_to_plot].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", square=True)
plt.title("Correlation Heatmap of Weather Variables")
plt.tight_layout()
plt.savefig("plots/correlation_heatmap_Ln.png")
plt.show()
```

## Feature Engineering

```python
# 1. Extract datetime features
hourly_df["hour"] = hourly_df["date"].dt.hour
hourly_df["dayofweek"] = hourly_df["date"].dt.dayofweek

# 2. Choose target variable
target_col = "temperature_2m"

# 3. Create sliding window data
def create_sequences(data, window_size=6):
    X, y = [], []
    for i in range(len(data) - window_size):
        X.append(data[i:i + window_size])
        y.append(data[i + window_size])
    return np.array(X), np.array(y)

# Use selected features (can be more)
features = ["temperature_2m", "hour", "dayofweek"]
data = hourly_df[features].dropna().reset_index(drop=True)

# Normalize if needed (optional now)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# Create sequences
X, y = create_sequences(data_scaled[:, 0], window_size=6)  # Using temp only for now

print(f"Shape of X: {X.shape}")
print(f"Shape of y: {y.shape}")
```

## Reshaping x value for preprocessing update from LSTM

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Reshape X and y
X = np.array(X)
y = np.array(y)

# Scale the data (LSTM performs better with scaled data)
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
y_scaled = scaler.fit_transform(y.reshape(-1, 1))

# Reshape X to [samples, timesteps, features]
X_scaled = X_scaled.reshape((X_scaled.shape[0], X_scaled.shape[1], 1))

print("✅ Data ready for LSTM:")
print("X shape:", X_scaled.shape)
print("y shape:", y_scaled.shape)
```

## Build and Train LSTM model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import EarlyStopping

# Define LSTM model
model = Sequential([
    LSTM(64, activation='relu', input_shape=(6, 1)),
    Dense(1)
])

# Compile model
model.compile(optimizer='adam', loss='mse')

# Optional: Add early stopping to prevent overfitting
early_stop = EarlyStopping(monitor='loss', patience=5)

# Train model
history = model.fit(X_scaled, y_scaled, epochs=50, batch_size=16, verbose=1, callbacks=[early_stop])

# Save model
model.save("lstm_temperature_model_Ln.h5")
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("LSTM Training Loss")
plt.legend()
plt.tight_layout()
plt.savefig("lstm_training_loss_Ln.png")
plt.show()
```

## Inverse Scaling

```python
# Predict on training data
y_pred_scaled = model.predict(X_scaled)

# Inverse transform predictions and labels
from joblib import dump, load
dump(scaler, "temp_scaler_Ln.pkl")  # save scaler for future use

y_pred = scaler.inverse_transform(y_pred_scaled)
y_actual = scaler.inverse_transform(y_scaled)

# Compare sample predictions
for i in range(5):
    print(f"Actual: {y_actual[i][0]:.2f} °C, Predicted: {y_pred[i][0]:.2f} °C")
```

```python
import matplotlib.pyplot as plt

X = X.reshape((X.shape[0], X.shape[1], 1))
preds = model.predict(X)
preds_inverse= scaler.inverse_transform(preds)
y_inverse = scaler.inverse_transform(y.reshape(-1, 1))

plt.figure(figsize= (12,6))
plt.plot(y_inverse, label="Actual", marker= 'o', linestyle= '-')
plt.plot(preds_inverse, label="Predicted", marker= 'x', linestyle= '--')
plt.title("Actual vs Predicted Temperature")
plt.xlabel("Time steps")
plt.ylabel("Temperature (°C)")
plt.legend()
plt.tight_layout()
plt.savefig("Actual_vs_pred_Ln.png")
plt.show()
```

## Metrics Evaluation

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

mae = mean_absolute_error(y_inverse, preds_inverse)
rmse = np.sqrt(mean_squared_error(y_inverse, preds_inverse))
r2 = r2_score(y_inverse, preds_inverse)

print(f"MAE: {mae:.2f} °C")
print(f"RMSE: {rmse:.2f} °C")
print(f"R² Score: {r2:.2f}")
```
